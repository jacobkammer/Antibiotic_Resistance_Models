```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Define the time-dependent parameter function
def p(t):
    # Example: parameter decreases over time as an exponential decay
    return 2 * np.exp(-0.1 * t)  # You can change this function as needed

# Define the differential equation
def dydt(y, t):
    return -p(t) * y

# Initial condition
y0 = 1

# Time points where we want the solution
t = np.linspace(0, 50, 500)

# Solve the differential equation
solution = odeint(dydt, y0, t)

# Plot the results
plt.plot(t, solution, label='y(t)')
plt.xlabel('Time')
plt.ylabel('y(t)')
plt.title('Solution of the Differential Equation with Time-Varying Parameter')
plt.grid()
plt.legend()
plt.show()
```

## The interp function in Python, specifically from the numpy library (numpy.interp), is used for one-dimensional linear interpolation. It allows you to estimate intermediate values within a given range based on known data points. This is particularly useful when you have a set of discrete data points and you want to find a value between them.¶
## How numpy.interp Works
## The function numpy.interp(x, xp, fp) takes the following parameters:
# x: The x-coordinates at which you want to interpolate.
# xp: The x-coordinates of the data points you know (must be in ascending order).
# fp: The y-coordinates of the data points you know, corresponding to xp.


## interp function length of all 3 vectors must be same otherwise get value error returned
